```python
import json
```

```python
with open('output.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        state_transfer = ""
        data = json.loads(line)
        record = data['record']
        for state in record:
            current_state = state['state']
            state_transfer += f"{current_state} -->"

state_transfer
```
