```python
import json
```

一个正常的RAG流程可由下面的正则语言表达：
$$P=((RAM)^*(RA(G|MI)))$$

$$Pipeline = D(P^+(\epsilon |E)P^*)C$$

理论上应该是$$P=((RAM)^*(RA(G|I)))$$但是我代码实现导致I前面必须有M，这在工程上容易解决，也无关系统的正确性

```python
def judge(meta_plan: str):
    if "Draft Plan:" in meta_plan:
        return True
    else:
        return False
```

```python
#获取每个数据的RAG执行序列
languages = []
map = {
    "assess": "RA",
    "refine": "M",
    "rag_generate": "G",
    "internal_generate": "I"    
}

with open("output.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        series = ""
        logs = data["logs"]
        for log in logs:
            if "meta_plan" in log:
                meta_plan = log['meta_plan']
                has_draft = judge(meta_plan)
                if has_draft:
                    series += "D"
            elif "meta_state" in log:
                if log["meta_state"] == "search":
                    # series += str("R")
                    logs = log.get('logs', '')
                    records = logs.get("records", None)
                    for record in records:
                        state = record['state']
                        series += str(map[state])
                elif log['meta_state'] == "reason":
                    series += str("E")
                elif log['meta_state'] == 'conclude':
                    series += "C"
        print(f"Series:{series}")
        languages.append(series)
```

D: Draft-Plan

R: Retrieval

E: Reason

C: Conclusion

A: Assess

M: Refine

I: Internal Generate

G: RAG Generate

```python
import re
p_pattern = r"(RAM)*RA(G|MI)"
full_regex = f"^D(?:{p_pattern})+E?(?:{p_pattern})*C$"

check_list = []
for pipeline in languages:
    match = re.fullmatch(full_regex, pipeline)
    if match:
        result = {"pipeline": pipeline, "check": True}
        check_list.append(result)
        print(result)
    else:
        result = {"pipeline": pipeline, "check": False}
        check_list.append(result)
        print(result)
```

```python
# 分析匹配失败的Case
fail_parse = []
for i, check in enumerate(check_list,1):
    if check['check'] == False:
        a = {"i": i, "pipeline": check['pipeline'], "check": check["check"]}
        fail_parse.append(a)
        print(a)
```
