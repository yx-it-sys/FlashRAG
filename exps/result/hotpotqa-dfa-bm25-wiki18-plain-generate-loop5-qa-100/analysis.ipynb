# 比较利用LLM内部先验的占比

- 对于OmniSearch，比较当前数据有没有retrieved documents
- 对于DFA，比较是否出现`S_Plain_Generate`这个状态

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# DFA-RAG
with open('output.jsonl', 'r', encoding='utf-8')as f:
    times = 0
    for line in f:
        data = json.loads(line)
        records = data['record']
        for record in records:
            state = record['state']
            if state == "plain_generate":
                times += 1
    ratio = times / 100
    print(ratio)
```
