# 比较利用LLM内部先验的占比

- 对于OmniSearch，比较当前数据有没有出现"role": "user"
- 对于DFA，比较是否出现`S_Plain_Generate`这个状态

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# OmniSearch
with open('output.jsonl', 'r', encoding='utf-8')as f:
    external = 0
    all_rounds = 0
    for line in f:
        data = json.loads(line)
        context = data['context']
        item_external = 0
        item_rounds = 0
        for item in context:
            item_rounds += 1
            if item.get('role') == 'user':
                item_external += 1
        external += item_external
        all_rounds += item_rounds
    internal_ratio = 1 - (external / all_rounds)

internal_ratio
```
